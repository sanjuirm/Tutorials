```python
import pandas as pd 
```

```python
# Reading a MultiIndex data
df = pd.read_csv('Dataset/MultiIndex.csv', index_col= [0,1,2], skiprows= [2], header= [0,1]).sort_index()
df
```

```python
# Lets see Row Indexes(Index) and column Coulmn Indexes

display(df.columns)
display(df.index)
```

Both are MultiIndex.  

```python
# Lets see Index and columns lables
display(df.columns.values)
display(df.index.values)
```

Here `('2010', 'Jan-Jun')` is the first column label(0th) and so on.

Similarly, `('CA Warehouses', 'Los Angeles', 'AC')` is the first row label(0th).

```python
# Lets get levels of Index and Columns
print('{} Index levels are:'.format(len(df.index.levels)))
display(df.index.levels)

print('{} column levels are:'.format(len(df.columns.levels)))
display(df.columns.levels)
```

```python
# Lets get level names of Index and Columns
print(df.columns.names,'\n',df.index.names)
```

Both Index and Column levels have no name

```python
# Lets assign names to the levels
df.index.names = ['State', 'Location', 'Product']
df.columns.names= ['Year', 'Quater']

print(df.columns.names,'\n',df.index.names)
```

Now we can see the both Index and Column Level names

```python
print(df.columns.levels[0].name)
print(df.index.levels[1].name)
```

Respective names of first column and second index name

```python
# Lets see an index level and its values
df.index.get_level_values(1) 

# equivalent to df.index.level[1].values
```

```python
# Simmilarly lets get a column level and its values
df.columns.get_level_values(1)

# equivalent to df.columns.level[1].values
```

#### SUMMARY
   Thus this MultiIndex DataFrame have 3 Index levels named `'State'`, `'Location'`, `'Product'`  and 2 Column index levels named `'Year'`, `'Quater'`. They are also called as `index names` and `column names`.
   
   Except this, first row label is `('CA Warehouses', 'Los Angeles', 'AC')` and so on. Similarly first column label is `('2010', 'Jan-Jun')`. 
   
   An axis may have different levels. In this DataFrame, rows, or 0th axis have three levels: `'State'`, `'Location'`, `'Product'`. For axis=1,Columns, the levels are `'Year'`, `'Quater'`.
