**1.** `pivot()` and `pivot_table()` both are used for reshaping a datframe(from long format to wide format).
   
**2.** `pivot()` doesn't support duplicate values but `pivot_table()` does. Hence `pivot()` doesn't support agrregation of data whereas `pivot_table()` does.    

```python
import pandas as pd
temp_df = pd.read_csv('Dataset/volcano_data_2010.csv')

pd.options.display.max_columns = len(temp_df.columns)
pd.options.display.max_rows = len(temp_df.index)

temp_df
```

```python
df = temp_df.loc[ : , ['Year', 'Name', 'Location', 'Country', 'Type', ]]
df
```

#### 1. `pivot_table()`

```python
df.pivot_table(values= 'Location', index='Country', columns= 'Year', aggfunc= 'count', fill_value= 0)
```

Lets understand what is happening. 

we are grouping by both of our index argument `Country` and column argument `Year`. Below We grouped the data by country `Indonesia` and below that, we can see that on Grouping the 'grouped data' by `Year` argument `Indonesia saw total 4 eruptions on year 2010`. 

Aggregation method opoerates on the `values` argument we pass. Here, on `Location` column `count` method is working. 

As on 2010, in Indonesia, eruption took place on 4 places, thus `count` method returns 4. This value equals the value in above Pivot table. Had `value_counts` been used, the result would have been diffrent.(Applied below)    

Thus behind the scences, the arguments we passed to `values`, `index` and `columns`parameters, aggregation and Group by takes place.  

```python
g = df.groupby('Country').get_group('Indonesia')
g
```

```python
g.groupby('Year').get_group(2010)
```

```python
df.pivot_table( values= 'Location', index= ['Type','Country'], columns= 'Year', aggfunc= 'value_counts', fill_value= 0)
```

Similar thing is happening here, aggregation and Group by. The diffrenece is the aggregation function, which is `value_counts`. 

As it calculates the total no of unique values in an axis, thus we get an extra index named `Location`. 

Before we were not getting it becoz, as `count` aggregation was applied on the `Location` column which simply returns the no. of non-na values. 

Thus we can see that in below df, on `Year` 2010, in `Indonesia`, among `Stratovolcano` group, 2 volacanic eruption took place in `Location` called Java, 1 in Sumatra, 1 in Sangihe Is-Indonesia. It matches with the value for the respective indexes and columns in the above pivot table. 

```python
g.groupby('Year').get_group(2010)
```

#### 2. `pivot()` 

No Aggregation possible

```python
# If we apply the same arguments here, it would give us Value error
df.pivot(values= 'Location', index='Country', columns= 'Year')
```

The reason this is happening becoz of presence of similar values in `country` column. 

As per offical documentation, 

>Uses unique values from specified index / columns to form axes of the resulting DataFrame

>This function does not support data aggregation

>Raises
ValueError:
When there are any index, columns combinations with multiple values. 

What happening is that, as there are many `Indonesia` in index axis and each have diffrent value as per location column( becoz eruption happened in diffrent places during year 2010 in Indonesia, like in java and sumatra), thus it raises the error due to failure to assign a value.

For example, in year 2010, index value `Indonesia` have multiple eruption. Hence the `ValueError`.   

```python
# Now if we remove the index parameter, it won't give us an error
df.pivot(values= 'Country', columns= 'Year')
```

Here on Grouping `year` column, each grouped year shows the country where eruption took place on a particular year. 

On 2010, we can see eruption took place in total 9 places (inclusive of similar places), which equals the pivot dataframe data above.

Now here no `ValueError` occurs becoz, **we have no defined index**. we are using the defualt 0,1.. index. **Thus no similar index, columns combination with multiple values occurs**.( which happens in the above discussion) 

For example, index of Ecuador in 2010 year is 0. we won't find country in the same year (or other) having similar index. even for year 2011, indexx starts from the end index vaue of year 2010. Thus there is no duplicate index.   

```python
display(df.groupby('Year').get_group(2010))
display(df.groupby('Year').get_group(2011))
```

**NOTE:** Remember it all depends upon the structure of index(es) or possible duplicated entries from the index of interest after grouping.

```python
df = pd.DataFrame({'foo': ['one', 'one', 'one', 'two', 'two',
                           'two'],
                   'bar': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'baz': [1, 2, 3, 4, 5, 6],
                   'zoo': ['x', 'y', 'z', 'q', 'w', 't']})
df
```

```python
df.pivot(index='foo', columns='bar', values='baz')
```

It is not showing any error, bcz of lack of similar index, column combinations with multiple values.
