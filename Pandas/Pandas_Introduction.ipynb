**Filtering a DataFrame**

```python
# using &(and) operator
filt = (df['University'] == 'Yes, full-time') & (df['MajorUndergrad'] == 'Computer science or software engineering')
df.loc[filt,['FormalEducation']]
```

```python
# using ~(not) operator. 
filt = df['Country'] == 'United Kingdom' # to filter dataframe using opposite of what we set as filter  
df.loc[~filt, 'Country']
```

```python
# using isin operator
countries= ['india', 'United Kingdom', 'France']
filt = df['Country'].isin(countries)
df.loc[filt,'Country']
```

```python
# searching whether a string contains a particualr value or not
filt = df['DeveloperType'].str.contains('Web developer', 'Mobile developer', na= False)
df.loc[filt,  'DeveloperType']
```

**Adding/Removing rows and columns**

```python
df1 = pd.DataFrame({'first': ['Corey', 'sanju', 'ponju'],
                    'last': ['Schafer', 'Doe', 'Halder'],
                    'email': ['sanjuirm98@gmail.com', 'work.sanju98@gmail.com', 'sanjuknight17@gmail.com']})
df1
```

```python
# adding columns
df1['first'] + ' ' + df1['last']
```

```python
# creating a new column by adding
df1['full_name'] = df1['first'] + ' ' + df1['last']
df1
```

```python
# deleting column
df1.drop(columns= ['first', 'last'], inplace= True)
df1 
```

```python
# spliting column
# expand() method
df1[['first','last']] =  df1['full_name'].str.split(' ', expand= True) # (expand =True) argument separates the list of strngs into separate columns
# mapping those separate columns 
df1
```

```python
df1.append({'first':'ton'}, ignore_index= True) # it has been removed from pandas
```

```python
# using concat() method instead
# adding a row into a dataframe
# ignore_index allows pandas to change default index no and assign new index no
df2 = pd.DataFrame({'first': 'Tony'}, index= [0]) # creating a dataframe then adding
df3 = pd.concat([df1,df2], ignore_index= True)
df3
```

```python
# adding along column or axis =1
pd.concat([df1,df2], ignore_index= True, axis=1)
```

```python
# removing rows by specyfying index no
df3.drop(index= 3)
```

**Sorting data**

```python
# sorting data
df3.sort_values(by= ['email'], ascending= True) # sorting based on alphabatical (ascending) orders on email column
```

```python
# sorting multiple columns.
# one column in ascending order another in descending order.
# pass True False parameter in list in ascending argument.
# True for email column, false for last column. order wise. 
df3.sort_values(by= ['email', 'last'], ascending= [True, False]) 
```

```python
# to sort by indexes
df3.sort_index()
```
