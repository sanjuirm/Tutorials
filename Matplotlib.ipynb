```python
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np

plt.style.use('ggplot')
```

```python
# Creating a simple Plot
x = np.linspace( 0, 4*np.pi, 1000)
y = np.sin(x)

# Parameters of plot() for line styling
plt.plot(x,y, linestyle = 'dotted', color = 'darkolivegreen', linewidth = 2, marker= ',') 

plt.xlabel('X-Axis')
plt.ylabel(None)   # skipping Y-Axis label
plt.title('Sine Wave')

plt.show()
```

```python
# Multiple Lineplots in a single subplot
fig, ax  = plt.subplots()

ax.plot([0,4,8,12], [6,8,4,2]) # ([x],[y])
ax.plot([0,3,9,12,15,18], [2,1,9,6,4,3])

ax.set_title('Multiple Lineplots in a single Subplot')
ax.legend(['1st Plot', '2nd Plot'], bbox_to_anchor =(0.99, 0.12), ncol = 2)  # Legend 

plt.tight_layout()
plt.show()
```

**LEGEND PLACEMENT**

`bbox_to_anchor` - location of the legend anchor w.r.t the plot bounding box. Anchor location is specified as (x,y)

`loc` - Location of the legend anchor w.r.t legend box. Default is top right.

**'legend anchor'** is the top left point of the legend box

```python
fig, ax  = plt.subplots()

ax.plot([0,4,8,12], [6,8,4,2], label = 'Line Plot')
ax.plot([0,3,9,12,15,18], [2,1,9,6,4,3], label = '2nd Plot')

# Changing legend placement
ax.legend( bbox_to_anchor = (0,-0.1), loc = 'upper left', ncol = 2) # (-0.1 to place the legend box at the bottom of the X-Axis)
```

```python
fig2, ax  = plt.subplots()

x1 = [0,4,8,12]
y1 = [6,8,4,2]

x2 = [0,3,9,12,15,18]
y2 = [2,1,9,6,4,3]

# Adding lables in the plot function itself as parameters
ax.plot(x1, y1, label = '1st Plot', linewidth = 2, marker = '*') # ([x],[y])
ax.plot(x2, y2, linestyle = '--', label = '2nd Plot')

ax.legend()

ax.set_title('Multiple Lineplots in a single Subplot')

plt.tight_layout()
plt.show()
```

**GRIDSPEC**

```python
# Gridspec
fig = plt.figure(figsize = (10,10), constrained_layout = True)  # 1. create a figure
fig.suptitle('Uneven subplots (Gridspec)')

#2. create a grid spec for 3 rows and 3 columns on that figure
gs = fig.add_gridspec(3,3)

#create a subplot in 0th row but occupying all the columns
ax1 = fig.add_subplot( gs[0, : ])  # we are using add_subplot on the figure handle and getting the subplot axis handle in return
ax1.set_facecolor('#9debd9')           
ax1.plot([7,8,9],[2,8,6], color= 'red', linewidth = 2, linestyle= 'dotted')

ax2 = fig.add_subplot( gs [ 1, :-1])
ax2.set_facecolor('#bedea0')

ax3 = fig.add_subplot( gs[1: , -1])
ax3.set_facecolor('#c8c8db')

ax4 = fig.add_subplot( gs[2,0] )
ax4.set_facecolor('#e6714e')

ax5 = fig.add_subplot( gs[2,1] )

plt.show()
```

```python
# Plotting inside of an Axes

fig = plt.figure( figsize = (7,5), constrained_layout  = True)
ax = fig.add_subplot()

ax.plot([7,8,9],[2,8,6])

# create an inset axes using inset_axes()
# the input array specifies the sizing and position of the inset axes
# in the form of lower-left corner coordinates of inset axes, and its width and height respectively as a fraction of original axes
axins = ax.inset_axes( [0.45, 0.12, 0.47, 0.35] )  

axins.plot([7,8,9],[2,8,6], color ='red')
axins.set_facecolor('#bedea0')

plt.show()
```

```python
# Centering axes of a plot
x = [-5, -4, 5, 8]
y = [-6, 5, 8,-10]

fig, ax = plt.subplots()
ax.plot(x,y)

# Removing the Spines
ax.spines['top'].set_color(None)
ax.spines['right'].set_color(None)

# Positioning the remaining Spines
ax.spines['bottom'].set_position('center')  # Instead of positing the intersection at (0,0), putting at center
ax.spines['left'].set_position('center')
```

```python
# Setting limit for x and y axis ticks

x = [0,4,8,12]
y = [6,8,4,2]

fig, ax = plt.subplots()
ax.plot(x,y)

# setting X-Axis and Y-Axis tick limit
ax.set_xlim(0, 15)
ax.set_ylim(0, 15)

plt.show()
```

**AXIS TICK POSITIONS AND TICK LABELS**

```python
# Setting axis tick positions and tick labels 
x = [-5, -4, 5, 8]
y = [-6, 5, 8,-10]

fig, ax = plt.subplots()
ax.plot(x,y)

# Tick position
ax.set_xticks([-8,-4,0,4,8,12])
ax.set_yticks([-12,-8,-4,0,4,8,12])

# Tick labels
ax.set_xticklabels( ['- eight','- four','zero','four', 'eight', 'twelve'])
ax.set_yticklabels([])  # passing empty y_ticks labels (ticks pos are still there)
```

```python
# Setting axis tick positions and tick labels 
x = [-5, -4, 5, 8]
y = [-6, 5, 8,-10]

fig, ax = plt.subplots()
ax.plot(x,y)

# Avioding ticks by passing an empty array
ax.set_xticks([])
ax.set_yticks([])
```

**Control spacing between ticks**

```python
# using MultipleLocator()

x = [0,1,2,3,4,5,6,7,8]
y = [8,6,4,2,9,7,6,3,1]

fig, ax = plt.subplots()
ax.plot(x,y)

ax.xaxis.set_major_locator( plt.MultipleLocator(4))  # Kepping major X-Axis ticks label as multiple of 4
ax.xaxis.set_minor_locator( plt.MultipleLocator(2))  # Kepping minor X-Axis ticks label as multiple of 2

ax.yaxis.set_major_locator( plt.MultipleLocator(3)) 
```

```python
# using IndexLocator()
# args = 'base': increment, 'offset': starting point

x = [0,1,2,3,4,5,6,7,8]
y = [8,6,4,2,9,7,6,3,1]

fig, ax = plt.subplots(constrained_layout = True)
ax.plot(x,y)

ax.xaxis.set_major_locator( plt.IndexLocator( base = 2.5, offset= 0.10))
ax.yaxis.set_major_locator( plt.IndexLocator( base = 1, offset= 0.50))
```

```python
# using LinearLocator
# args = 'numticks' : no. of ticks

x = [0,1,2,3,4,5,6,7,8]
y = [8,6,4,2,9,7,6,3,1]

fig, ax = plt.subplots(constrained_layout = True)
ax.plot(x,y)

ax.xaxis.set_major_locator( plt.LinearLocator( numticks = 10))
ax.yaxis.set_major_locator( plt.LinearLocator( numticks = 6) )
```

```python
# using MaxNLocator
# args = 'n': no. of ticks, 'integer' : bool value (True to get only integer tick labels)

x = [0,1,2,3,4,5,6,7,8]
y = [8,6,4,2,9,7,6,3,1]

fig, ax = plt.subplots(constrained_layout = True)
ax.plot(x,y)

ax.xaxis.set_major_locator( plt.MaxNLocator(5))
ax.yaxis.set_major_locator( plt.MaxNLocator(6, integer = True) )
```

```python
# LogLocator for logarithmic ticks
# args = 'base': base level for log

x = [0,20,300,50000,600000,7000000]
y = [8,7,6,3,5,9]

fig, ax = plt.subplots(constrained_layout = True)
ax.plot(x,y)

# make axis scale as 'log'
ax.set_xscale('log')

ax.xaxis.set_major_locator( plt.LogLocator(base = 10))
```
